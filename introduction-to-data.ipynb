```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the image and labels
images = np.load("./cs475_project_data/images.npy")
labels = np.load("./cs475_project_data/labels.npy")
```

```python
# separate out class 0 and visualize the first image
class_0_images = images[labels == 0]
plt.imshow(class_0_images[0])
plt.set_cmap('gray')
```

```python
# for class 1
class_1_images = images[labels == 1]
plt.imshow(class_1_images[0])
plt.set_cmap('gray')
```

```python
# for class 2
class_2_images = images[labels == 2]
plt.imshow(class_2_images[0])
plt.set_cmap('gray')
```

```python
# for class 3
class_3_images = images[labels == 3]
plt.imshow(class_3_images[0])
plt.set_cmap('gray')
```

```python
# for class 4
class_4_images = images[labels == 4]
plt.imshow(class_4_images[0])
plt.set_cmap('gray')
```

```python
# Reshape the images data array from (num_images, height, width) to (num_images, height * weight)
input = images.copy()
num_imgs, height, width = input.shape
input = input.reshape(num_imgs, height*width) # shape each image to a vector, which is more often used
print(input.shape) # new shape of the images

# print the firts 5 flattened vectors (the first 5 images) with plt.plot
for i in range(0, 5):
    plt.figure(figsize=(20,6))
    plt.plot(input[i], 'r')
```
